```python
y1, y2, y3, z1, z2, z3 = symbols('y1, y2, y3, z1, z2, z3')
```

```python

ks = []
#vs = [0.9878042228*v[0] + 0.6802782372*v[-3], 0.9923751319*v[0] + 0.4964510255*v[-3],0.9874137376*v[0] + 0.6948766460*v[-3],0.3003651625*v[-3] - 0.00716082355*v[0], 0.4906983058*v[-3] - 0.00522580027*v[0],   0.2852226010*v[-3]- 0.00731449101*v[0]]
                    
vs = []
#vs =[1.782741376,1.728868443,1.829729471,-0.01598169399,0.2100827025,-0.007342470031]

#vs = [1.66808246, 1.4888261574, 1.6822903836, 0.29320433895, 0.48547250553, 0.27790810999]

def linear_equation(c1, c2, c3, c4, c5, c6):
    
    eq1 =sp.Eq(-0.2949582188e-1*y3+1.147055149*z3+0.51594241e-3*y2-y2+2.427688399*z2-0.2148538728e-2*y1+.4770180184*z1 , -c1)
    eq2 =sp.Eq(-0.5160018172e-2*y3-y3+1.186217417*z3+0.6192921461e-3*y2+0.6680655564e-1*z2-0.1586011615e-3*y1+0.3756480678e-1*z1 ,-c2)

    eq3 =sp.Eq(-0.3017149629e-1*y3+1.085604910*z3-0.1181792169e-1*y2+2.903389798*z2-0.7424319581e-2*y1-y1+1.206128646*z1 , -c3)

    eq4 =sp.Eq(-0.1207426474e-1*y3-1.188625236*z3-0.2555461473e-1*y2-2.452727072*z2-z2-0.5021242299e-2*y1-.4841877994*z1 , -c4)

    eq5 =sp.Eq(-0.1248649912e-1*y3-1.203863934*z3-z3-0.7032269015e-3*y2-0.6689049039e-1*z2-0.3954190186e-3*y1-0.3811882695e-1*z1 ,-c5)

    eq6 =sp.Eq(-0.1142742010e-1*y3-1.127203826*z3-0.3056199788e-1*y2-2.945769718*z2-0.1269609101e-1*y1-1.226249056*z1-z1 ,  -c6)


    ans = sp.solve((eq1, eq2, eq3, eq4, eq5, eq6), (y1,y3, y2, z1,z3,z2))
    for k, v in ans.items():
        v = N((v),10)
        vs.append(v)
        ks.append(k)
        

linear_equation(1.66808246, 1.4888261574, 1.6822903836, 0.29320433895, 0.48547250553, 0.27790810999)
vs
```
